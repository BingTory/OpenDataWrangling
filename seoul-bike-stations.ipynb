# 따릉이 대여소 정보
* https://www.bikeseoul.com/
* 따릉이 홈페이지에 나와있는 대여소 정보를 가져온다.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import requests
import pandas as pd
```

```python
url = 'https://www.bikeseoul.com/commonCode/getStationName.do'
 
%time response = requests.post(url=url)
response
```

```python
data = response.json()
```

```python
df = pd.DataFrame(data['stationList'])
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
stations = df[['lang', 'stationId', 'stationName']]
stations.head()
```

```python
stations['id'] = df['stationName'].apply(lambda x : x.split('.')[0] if '.' in x else x )
stations['name'] = df['stationName'].apply(lambda x : x.split('.')[1] if '.' in x else x )
stations.head()
```

```python
get_station_url = 'https://www.bikeseoul.com/commonCode/getStation.do'
 
%time response = requests.post(url=get_station_url)
response
```

```python
data = response.json()
stations = pd.DataFrame(data['stationList'])
stations
```
