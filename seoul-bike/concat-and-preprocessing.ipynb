```python
import pandas as pd
import numpy as np
```

```python
%ls data/
```

```python
import os

def concat_data(year):
    df_lists = []
    folder_path = "data/"
    for root, dirs, files in os.walk(folder_path):  
        for filename in files:
            if filename.startswith('서울특별시 공공자전거 대여이력 정보_{}년'.format(year)) :
                df_temp = pd.read_csv(folder_path+filename, encoding='cp949')
                print('-'*20)
                print(filename)
                print(df_temp.shape)
                df_lists.append(df_temp)
    df = pd.concat(df_lists)
    print(df.shape)
    return df        
```

```python
%time df_2016 = concat_data(2016)
print(df_2016.shape)
```

```python
%time df_2017 = concat_data(2017)
print(df_2017.shape)
```

```python
%time df_2018 = concat_data(2018)
print(df_2018.shape)
```

```python
def preprocessing(df):
    # 컬럼에 있는 '를 제거한다.
    df.columns = df.columns.str.strip("'")
    # 불필요하게 '가 들어가 있다. 스트링 타입으로 되어있는 데이터에서 '을 제거하자.
    %time df = df.apply(lambda x: x.str.strip("'") if x.dtype == np.dtype('object') else x)
    return df
```

```python
years = [2016, 2017, 2018]
for year in years:
    df = eval('df_'+str(year))
    %time df = preprocessing(df)
    df.to_csv('data/seoul-bike-{}.csv'.format(year), index=False)   
    print(df.shape)
```
